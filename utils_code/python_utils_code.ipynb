```python
import pandas as pd
import os
```

```python
current_path = os.getcwd()
covid_death_data_set = 'data_sets/NashvilleHousingData.xlsx'
current_path
```

```python
covid_deaths_data = os.path.join(current_path,covid_death_data_set)
covid_deaths_data
```

```python
df = pd.read_excel(covid_deaths_data)
```

```python
df.head(10)
```

```python
with open('NashvilleHousingData.csv', 'w') as csv_file:
    df.to_csv(path_or_buf=csv_file,index=False)
```

```python
def import_excel_to_csv(filename,dataframe):
    with open(filename, 'w') as csv_file:
        dataframe.to_csv(path_or_buf=csv_file,index=False)
```

```python
import_excel_to_csv(filename='NashvilleHousingData.csv',dataframe=df)
```
